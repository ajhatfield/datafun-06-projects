### Loading in a text file 'basketball.txt'

```python
from pathlib import Path

text = Path('basketball.txt').read_text()
```

### Setting the shape of the wordcloud to a circle or circle.png

```python
import imageio

mask_image = imageio.imread('circle.png')
```

### configuring the WordCloud object

```python
from wordcloud import WordCloud

wordcloud = WordCloud(colormap= 'prism', mask=mask_image, background_color='white')
```

### creating the word cloud 

```python
wordcloud = wordcloud.generate(text)
```

### Make new file called 'basketball.png' with word cloud

```python
wordcloud = wordcloud.to_file('basketball.png')
```

### display word cloud in Python

```python
import matplotlib.pyplot as plt
plt.imshow(wordcloud)
```
